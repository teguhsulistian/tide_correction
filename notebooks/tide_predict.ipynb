```python
import pandas as pd
import numpy as np
from eo_tides.model import model_tides
```

```python
tide_dir=r"D:\tide_models"
model="EOT20"
x = 115.00,
y = 8.10,
start_time = "2021-08-29 3:22:00" # In UTC time, Start time for tide predictions (in the format "YYYY-MM-DD HH:MM:SS") 
end_time = "2021-08-29 3:56:00" # In UTC time, End time for tide predictions (in the format "YYYY-MM-DD HH:MM:SS") 
freq = "30min" # Frequency of tide predictions (e.g., '1H' for hourly) and '10min' for every 10 minutes
```

```python
 tide_results = model_tides(
        x=x,
        y=y,
        time= pd.date_range( start_time, end_time, freq=freq),
        model=model,
        directory=tide_dir
    )

tide_results
```
